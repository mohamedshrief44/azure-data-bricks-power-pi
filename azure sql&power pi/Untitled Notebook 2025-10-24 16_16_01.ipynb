```python
from pyspark.sql.functions import col
```

```python
df_customer = df_customer.select(
    col("customer_id").cast("int"),
    col("full_name"),
    col("email"),
    col("country"),
    col("registration_date").cast("date")
)
```

```python
joined = df_transaction.join(df_customer,"customer_id").join(df_product,"product_id").join(df_store,"store_id").withColumn("total_amount",col("quantity")*col("price"))
```

```python
joined.write.mode("overwrite").saveAsTable("joined")
```
